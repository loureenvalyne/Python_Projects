### Combining 50km/50miles with isin

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] ==2017)]
```

```python
df[df['Event name'] == 'Destin 50 Mile Beach Ultra (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] ==2017) & (df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'])]
```
